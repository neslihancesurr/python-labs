```python
### Quick EDA — Average charges by gender
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/insurance.csv")
df.info()
```

```python
charges_sex = df.groupby('sex')['charges'].mean().to_frame()
charges_sex.index = charges_sex.index.map({0: 'Male', 1: 'Female'})
print(charges_sex)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=charges_sex.reset_index(), x='sex', y='charges', hue='sex', palette='Set2', legend=False)
plt.title("Average Insurance Charges by Gender")
plt.xlabel("Sex")
plt.ylabel("Average Charges ($)")
plt.tight_layout()
plt.show()
```
